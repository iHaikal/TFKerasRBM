```
from sklearn.datasets import make_blobs

features = 6
samples = 10**4

data = make_blobs(n_samples=samples, centers=(features + 1), n_features=features)

x, y = data

x[0], y[0]
```

```
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
norm_data = scaler.fit_transform(x).astype('float32')

norm_data[0:3]
```

```
%matplotlib inline

import pandas as pd

from rbm import PredictedOutput, RBM

model = RBM(features, 2)

prob, _= model.predict(norm_data, predict_output=PredictedOutput.hidden)
reduced_df = pd.DataFrame(dict(x=prob[:, 0], y=prob[:, 1], label=y))

reduced_df.head(3)
```

```
import plotly.express as px

fig = px.scatter(reduced_df, x='x', y='y', color='label',template='seaborn')
fig.update_traces(marker=dict(size=3,opacity=.6))
```
